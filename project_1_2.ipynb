```python
class Square:
    def __init__(self):
        self.calculate_more = True  # 사용자가 더 계산할지 여부를 나타내는 플래그 설정

    def calculate_area(self, shape_type):
        def parse_input(input_str):
            # 사용자 입력에서 숫자를 추출하여 리스트로 반환하는 함수
            return [int(num) for num in input_str if num.isdigit()]

        if shape_type == '직사각형':
            input_str = input('가로, 세로를 입력하세요. 예시 : 가로,세로 또는 가로 세로\n >>>')
            input_list = parse_input(input_str)
            if len(input_list) == 1:
                width, vertical = input_list[0], 1  # 한 개 값만 들어온 경우 가로로 간주하고 세로는 1로 설정
            elif len(input_list) == 2:
                width, vertical = input_list
            else:
                return '잘못된 입력입니다.'
            area = width * vertical
            result = '직사각형의 넓이는 : ' + str(area)
        elif shape_type == '평행사변형':
            input_str = input('밑변, 높이를 입력하세요. 예시 : 밑변,높이 또는 밑변 높이\n >>>')
            input_list = parse_input(input_str)
            if len(input_list) == 1:
                base, height = input_list[0], 1  # 한 개 값만 들어온 경우 밑변으로 간주하고 높이는 1로 설정
            elif len(input_list) == 2:
                base, height = input_list
            else:
                return '잘못된 입력입니다.'
            area = base * height
            result = '평행사변형의 넓이는 : ' + str(area)
        elif shape_type == '사다리꼴':
            input_str = input('밑변1, 밑변2, 높이를 입력하세요. 예시 : 밑변1,밑변2,높이 또는 밑변1 밑변2 높이\n >>>')
            input_list = parse_input(input_str)
            if len(input_list) == 1:
                base1, base2, height = input_list[0], 1, 1  # 한 개 값만 들어온 경우 밑변으로 간주하고 나머지는 1로 설정
            elif len(input_list) == 3:
                base1, base2, height = input_list
            else:
                return '잘못된 입력입니다.'
            area = ((base1 + base2) * height) / 2
            result = '사다리꼴의 넓이는 : ' + str(area)
        else:
            result = '에쿠, 계산이 불가능한 도형이에요.'
        return result

    def ask_continue(self):
        answer = input("계속 사각형 넓이를 계산해볼까요? (네: 1 / 아니오: 2) : ")
        if answer.lower() in ['네', '1']:
            return True
        elif answer.lower() in ['아니오', '2']:
            return False
        else:
            print("올바른 값을 입력하세요.")
            return self.ask_continue()

    def select_shape(self):
        while self.calculate_more:  # 사용자가 더 계산하려고 할 때까지 반복
            square = int(input('이번에 넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

            if square == 1:
                print(self.calculate_area('직사각형'))
            elif square == 2:
                print(self.calculate_area('평행사변형'))
            elif square == 3:
                print(self.calculate_area('사다리꼴'))
            else:
                print("잘못된 입력이에요. 다시 시도해주세요.")

            self.calculate_more = self.ask_continue()  # 사용자에게 계속할지 묻고, 플래그 업데이트

# 객체 생성
a = Square()
# 도형 선택 및 넓이 계산
a.select_shape()

print("수학은 즐겁죠!")
```

```python
class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.order_menu = []  # 주문 리스트
        self.order_price = []  # 가격 리스트

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(f"{i+1}. {self.menu[i]} : {self.price[i]}원")

    # 주문 메서드
    def menu_select(self):
        print()  # 줄 바꿈
        while True:
            n = int(input("음료 번호를 입력하세요 : "))  # 음료 번호 입력

            # 메뉴판에 있는 음료 번호일 때
            if 1 <= n <= len(self.menu):
                self.order_price.append(self.price[n-1])  # 가격 리스트에 추가합니다.
                print(f"{self.menu[n-1]} : {self.price[n-1]}원")
            else:
                print("없는 메뉴이군요. 다시 주문 부탁드려요.")
                continue

            # 음료 온도 물어보기
            while True:
                t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                if t == 1:
                    self.order_menu.append("HOT " + self.menu[n-1])  # 주문 리스트에 추가합니다.
                    break
                elif t == 2:
                    self.order_menu.append("ICE " + self.menu[n-1])  # 주문 리스트에 추가합니다.
                    break
                else:
                    print("1과 2 중 하나를 입력하세요.")
                    continue

            # 추가 주문 여부 물어보기
            while True:
                additional_order = input("추가 주문을 하시겠습니까? (예: 1, 아니오: 2) : ")
                if additional_order == '1':
                    break
                elif additional_order == '2':
                    return
                else:
                    print("잘못된 입력이군요.")
                    continue

    # 결제 메서드
    def pay(self):
        while True:
            payment_method = input("결제 방법을 선택하세요 (현금: 1, 카드: 2) : ")
            if payment_method == "1" or payment_method == "2":
                if payment_method == "1":
                    print("현금 계산을 위해 직원을 불러드릴게요!")
                else:
                    print("카드 결제해드릴게요! IC카드를 올바른 방향으로 꽂아주세요.")
                break
            else:
                print("결제 방법을 다시 선택하셔야해요. (현금: 1, 카드: 2)")
                help_needed = input("직원 도움이 필요하신가요? (네: 3, 아니오: 4) : ")
                if help_needed == "3":
                    print("직원을 불러드리겠습니다! 잠시만 기다려주세요")
                elif help_needed == "4":
                    continue


    # 주문표 출력 메서드
    def table(self):
        # 외곽
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        # 주문 상품명 : 해당 금액
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i], "원")

        print('합계 금액 :', sum(self.order_price))

        # 외곽
        for i in range(5):
            print('|' + ' ' * 31+ '|')
        print('⟝' + '-' * 30 + '⟞')

# 키오스크 객체 생성
kiosk = Kiosk()

# 주문
kiosk.menu_print()
kiosk.menu_select()

# 결제
kiosk.pay()

# 주문표 출력
kiosk.table()

print("맛있게 드세요!")
```
