```python

# 메인 퀘스트 2번: 아이펠 통합 관리 프로그램
# 1. Aiffel 클래스 만들기:
# 학번은 generate_student_id 메서드를 통해 생성되며, 학번 생성 규칙을 따릅니다.
# 2. 점수 관리 메서드:
# add_score: 점수를 추가합니다.
# deduct_score: 점수를 차감합니다.
# get_total_score: 현재 총점을 반환합니다.
# 3. 그루 명단 관리하기:
# display_all_grus: 전체 명단과 정보를 출력합니다.
# sort_grus_by_score: 점수 기준으로 내림차순 정렬된 명단을 출력합니다.
# remove_gru: 특정 학번의 그루 정보를 삭제합니다.
# 4. 그룹핑 기능 추가하기:
# create_groups: 주어진 그룹 크기(group_size)로 랜덤하게 조를 구성하고 출력합니다.

import random

class Aiffel:
    # 전체 그루 명단을 저장하는 클래스 변수
    grus_list = []

    # 1. Aiffel 클래스 만들기
    def __init__(self, name, year, batch, course):
        # 이름, 입학연도, 기수, 과정을 받아 초기화
        self.name = name
        self.year = year
        self.batch = batch
        self.course = course
        # 학번 생성
        self.student_id = self.generate_student_id()
        self.scores = 0
        # 생성된 객체를 전체 그루 명단에 추가
        Aiffel.grus_list.append(self)

    def generate_student_id(self):
        # 연도 마지막 두자리
        YY = str(self.year)[-2:]
        # 기수는 두자리로 표현
        NN = f"{self.batch:02}"
        # 과정 종류에 따라 코드 지정 (코어 = 2, 리서치 = 1)
        C = '2' if self.course == "코어" else '1'
        # 가나다순 번호는 두자리로 표현
        XX = f"{len(Aiffel.grus_list) + 1:02}"
        return YY + NN + C + XX

    # 2. 점수 관리 메서드
    def add_score(self, points):
        # 점수 추가
        self.scores += points

    def deduct_score(self, points):
        # 점수 차감
        self.scores -= points

    def get_total_score(self):
        # 현재 총점 반환
        return self.scores

    # 3. 그루 명단 관리하기
    @classmethod
    def display_all_grus(cls):
        # 전체 명단과 정보 출력
        for gru in cls.grus_list:
            print(f"이름: {gru.name}, 학번: {gru.student_id}, 점수: {gru.scores}")

    @classmethod
    def sort_grus_by_score(cls):
        # 점수를 기준으로 내림차순 정렬
        sorted_list = sorted(cls.grus_list, key=lambda x: x.scores, reverse=True)
        for gru in sorted_list:
            print(f"이름: {gru.name}, 학번: {gru.student_id}, 점수: {gru.scores}")

    @classmethod
    def remove_gru(cls, student_id):
        # 특정 학번의 그루 정보 삭제
        for gru in cls.grus_list:
            if gru.student_id == student_id:
                cls.grus_list.remove(gru)
                print(f"학번 {student_id}의 {gru.name} 그루가 삭제되었습니다.")
                return
        print(f"학번 {student_id}의 그루를 찾을 수 없습니다.")

    # 4. 그룹핑 기능 추가하기
    @classmethod
    def create_groups(cls, group_size):
        # 주어진 그룹 크기로 랜덤하게 조 구성
        random.shuffle(cls.grus_list)
        groups = [cls.grus_list[i:i + group_size] for i in range(0, len(cls.grus_list), group_size)]
        for idx, group in enumerate(groups):
            print(f"\nGroup {idx + 1}:")
            for gru in group:
                print(f"이름: {gru.name}, 학번: {gru.student_id}, 점수: {gru.scores}")

# 예시 그루 생성
gru1 = Aiffel("홍길동", 2024, 31, "코어")
gru2 = Aiffel("김철수", 2024, 31, "코어")
gru3 = Aiffel("이영희", 2024, 31, "코어")
gru4 = Aiffel("박민수", 2024, 31, "코어")
gru5 = Aiffel("최지우", 2024, 31, "코어")

# 점수 부여 및 차감 예시
gru1.add_score(100)  # 홍길동에게 100점 부여
gru2.add_score(80)   # 김철수에게 80점 부여
gru3.add_score(90)   # 이영희에게 90점 부여
gru4.add_score(85)   # 박민수에게 85점 부여
gru5.add_score(95)   # 최지우에게 95점 부여

gru2.deduct_score(10)  # 김철수에게서 10점 차감
gru4.deduct_score(5)   # 박민수에게서 5점 차감

# 전체 명단과 정보 출력
print("\n전체 명단과 정보:")
Aiffel.display_all_grus()

# 점수 기준 내림차순 정렬
print("\n점수 기준 내림차순 정렬:")
Aiffel.sort_grus_by_score()

# 특정 인물 정보 삭제
print("\n특정 인물 정보 삭제:")
Aiffel.remove_gru("243122")

# 그룹핑 예시 (2인 조로 그룹핑)
print("\n2인 조로 그룹핑:")
Aiffel.create_groups(2)
```

```python
# 개선 내용 추가
# 1. 변경 내역에 날짜와 시간 기록: record_history 메서드를 추가하여 점수 변경 내역을 기록합니다.
# 2. 이름을 기준으로 정렬하여 학번의 가나다순 번호 유지: Aiffel.sort_grus_list 메서드를 통해 이름을 기준으로 정렬하여 학번의 가나다순 번호를 유지합니다.
# 3. 학번 중복 방지: 이미 존재하는 학번을 재사용하지 않도록 generate_student_id 메서드에서 학번을 중복되지 않게 보장합니다.
# 4. 없는 학번을 입력할 경우 에러 메시지 출력: remove_gru 메서드에서 없는 학번을 입력할 경우 에러 메시지를 출력합니다.
# 5. 점수 입력 유효성 검사: 점수 추가/차감 시 음수 값을 입력하지 못하도록 유효성 검사를 추가합니다.

from datetime import datetime  # 날짜와 시간을 다루기 위해 datetime 모듈을 import합니다.

class Aiffel:
    grus_list = []  # 코어 그룹의 그루들을 저장할 클래스 변수를 생성합니다.

    # Aiffel 클래스의 생성자입니다. 이름, 입학연도, 기수, 과정명을 입력받습니다.
    # 이를 통해 학번을 생성하고, 학번 정보로 코어 그룹의 그루들을 관리합니다.
    def __init__(self, name, admission_year, batch, course):
        self.name = name  # 이름을 인스턴스 변수로 저장합니다.
        self.admission_year = admission_year  # 입학연도를 인스턴스 변수로 저장합니다.
        self.batch = batch  # 기수를 인스턴스 변수로 저장합니다.
        self.course = course  # 과정명을 인스턴스 변수로 저장합니다.
        self.points = 0  # 퀘스트 점수를 초기화합니다.
        self.history = []  # 퀘스트 점수 변경 내역을 저장할 리스트를 초기화합니다.
        self.student_id = self.generate_student_id()  # 학번을 생성합니다.
        Aiffel.grus_list.append(self)  # 그루 객체를 코어 그룹의 그루들 리스트에 추가합니다.

    # 1. 변경 내역에 날짜와 시간 기록: record_history 메서드를 추가하여 점수 변경 내역을 기록합니다.
    def record_history(self, action, points):
        now = datetime.now()  # 현재 날짜와 시간 정보를 가져옵니다.
        formatted_date = now.strftime("%Y-%m-%d %H:%M:%S")  # 날짜와 시간을 지정된 형식으로 변환합니다.
        self.history.append((formatted_date, action, points))  # 변경 내역을 튜플로 생성하여 히스토리 리스트에 추가합니다.

    # 2. 이름을 기준으로 정렬하여 학번의 가나다순 번호 유지: Aiffel.sort_grus_list 메서드를 통해 이름을 기준으로 정렬하여 학번의 가나다순 번호를 유지합니다.
    @classmethod
    def sort_grus_list(cls):
        cls.grus_list.sort(key=lambda gru: gru.name)  # 이름을 기준으로 그루 객체 리스트를 정렬합니다.

    # 3. 학번 중복 방지: 이미 존재하는 학번을 재사용하지 않도록 generate_student_id 메서드에서 학번을 중복되지 않게 보장합니다.
    def generate_student_id(self):
        YY = str(self.admission_year)[-2:]  # 입학연도의 마지막 두 자리를 가져옵니다.
        NN = f"{self.batch:02}"  # 기수를 두 자리 숫자로 변환합니다.
        C = '2' if self.course == "코어" else '1'  # 코어인 경우 '2', 리서치인 경우 '1'을 지정합니다.
        index = Aiffel.grus_list.index(self)  # 현재 그루 객체의 인덱스를 가져옵니다.
        XX = f"{index + 1:02}"  # 인덱스에 1을 더한 값을 두 자리 숫자로 변환합니다.
        return YY + NN + C + XX  # 학번을 조합하여 반환합니다.

    # 4. 없는 학번을 입력할 경우 에러 메시지 출력: remove_gru 메서드에서 없는 학번을 입력할 경우 에러 메시지를 출력합니다.
    @classmethod
    def remove_gru(cls, student_id):
        gru_to_remove = None  # 삭제할 그루 객체를 저장할 변수를 초기화합니다.
        for gru in cls.grus_list:  # 코어 그룹의 그루들을 순회합니다.
            if gru.student_id == student_id:  # 입력한 학번과 일치하는 그루를 찾으면
                gru_to_remove = gru  # 삭제할 그루 객체를 저장합니다.
                break
        if gru_to_remove:  # 삭제할 그루 객체가 존재하면
            cls.grus_list.remove(gru_to_remove)  # 코어 그룹의 그루들 리스트에서 그루 객체를 삭제합니다.
        else:  # 삭제할 그루 객체가 존재하지 않으면
            print(f"학번 {student_id}의 그루를 찾을 수 없습니다.")  # 에러 메시지를 출력합니다.

    # 5. 점수 입력 유효성 검사: 점수 추가/차감 시 음수 값을 입력하지 못하도록 유효성 검사를 추가합니다.
    def add_points(self, points):
        if points < 0:  # 입력한 점수가 음수이면
            print("퀘스트 점수는 음수가 될 수 없습니다.")  # 에러 메시지를 출력합니다.
        else:  # 입력한 점수가 음수가 아니면
            self.points += points  # 퀘스트 점수를 추가합니다.
            self.record_history("add", points)  # 변경 내역을 기록합니다.

    # 점수를 차감하는 메서드입니다.
    def deduct_points(self, points):
        if points < 0:  # 입력한 점수가 음수이면
            print("페널티 점수는 음수가 될 수 없습니다.")  # 에러 메시지를 출력합니다.
        else:  # 입력한 점수가 음수가 아니면
            self.points -= points  # 퀘스트 점수를 차감합니다.
            self.record_history("deduct", points)  # 변경 내역을 기록합니다.

```
